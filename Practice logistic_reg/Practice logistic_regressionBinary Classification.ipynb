<h3 style="color:yellow" align="center">Predicting if a person would buy life insurnace based on his age using logistic regression</h3>

Above is a binary logistic regression problem as there are only two possible outcomes (i.e. if person buys insurance or he/she doesn't). 

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("insurance_data.csv")
df.head()
```

```python
df.shape
```

```python
plt.scatter(df.age,df.bought_insurance,marker='+',color='red')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['age']],df.bought_insurance,test_size=0.2)
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_predicted = model.predict(X_test)
```

```python
X_test
```

```python
y_predicted
```

```python
model.predict_proba(X_test)
```

```python
model.score(X_test,y_test)
```

```python
y_predicted
```

```python
X_test
```

**model.coef_ indicates value of m in y=m*x + b equation**

```python
model.coef_
```

**model.intercept_ indicates value of b in y=m*x + b equation**

```python
model.intercept_
```

**Lets defined sigmoid function now and do the math with hand**

```python
import math
def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
def prediction_function(age):
    z = model.coef_ * age + model.intercept_
    y = sigmoid(z)
    if y >=0.5:
        print(f"Prediction is 1, that is ersio going to take insurence with this {age}")
    else:
        print(f"Prediction is 0, that is ersion not going to take insurence with this {age}")
```

```python
age = 35
prediction_function(age)
```

```python
age = 45
prediction_function(age)
```

**0.26047 is less than 0.5 which means person with 35 age will *not* buy insurance**

```python
age = 43
prediction_function(age)
```

**0.5470 is more than 0.5 which means person with 43 will buy the insurance**
